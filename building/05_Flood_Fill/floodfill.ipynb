```python
import numpy as np
from collections import deque


def flood_fill(arr, start):
    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]
    
    queue = deque([start])
    visited = set([start])
    
    while queue:
        y, x = queue.popleft()
        if (
            x in range(arr.shape[1]) 
            and y in range(arr.shape[0]) 
            and arr[y, x] == " "
        ):
            arr[y, x] = "@"
            
            for dx, dy in directions:
                neighbor = (y + dy, x + dx)
                if neighbor not in visited:
                    queue.append(neighbor)
                    visited.add(neighbor)


start = (0, 0)

with open("floor.txt") as f:
    data = [list(row) for row in f.read().strip("\n").split("\n")]
h, w = len(data), len(data[0])

cave = np.array(data, dtype="<U1")

flood_fill(cave, start)
np.savetxt('flooranswer.txt', cave, fmt='%s', delimiter='')
print(np.count_nonzero(cave == '@'))
```
